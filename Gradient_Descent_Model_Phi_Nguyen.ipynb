**Gradient Descent Practice**

mathematical implementation

```python
import numpy as np
from matplotlib import pyplot as plt
import math
#data
x = 10 * np.random.rand(100, 1)
y = 4 + 2*x + 3*np.random.randn(100,1)

w = 0.0
b = 0.0

m = x.shape[0]
a = 0.005

xlist = []
ylist = []


def descent(x,y,w,b,a,m):
  jw = 0.0
  jb = 0.0

  for i in range(m):
    jw += x[i] * ((w*x[i] + b) - y[i])
    jb += ((w*x[i] + b) - y[i])

  w = w - a*jw*(1/m)
  b = b - a*jb*(1/m)

  return w, b

for epoch in range(2500):
  w, b = descent(x,y,w,b,a,m)

for i in range(math.floor(np.min(x)), math.ceil(np.max(x)) + 1):
  xlist.append(i)
  ylist.append(w*i + b)


plt.scatter(x, y)
plt.plot( xlist, ylist, 'r--')
plt.title('Visualization')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show
```

**Polynomial Regression**

using sklearn libraries

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

x = 4 * np.random.rand(100, 1)
y = 4 + 2*x +12*x**2 + 8*x**3+ 50*np.random.randn(100,1)

pf = PolynomialFeatures(degree = 4, include_bias = False)
xp = pf.fit_transform(x)

reg = LinearRegression()
reg.fit(xp, y)

xval = np.linspace(0,4,100).reshape(-1,1)
xvp = pf.transform(xval)
yval = reg.predict(xvp)

plt.scatter(x, y)
plt.plot(xval, yval, 'r--')
plt.show

```
